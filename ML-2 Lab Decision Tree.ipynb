```python
grade = []
for i in df['Total\n(out of 100)'].values:
    if i>= 0 and i< 40:
        grade.append('F')
    elif i>= 40 and i< 60:
        grade.append('P')
    elif i>= 60 and i< 80:
        grade.append('A')
    elif i>= 80 and i< 90:
        grade.append('G')
    else:
        grade.append('O')


df_Copy = df
se = pd.Series(grade)
df_Copy['Grade'] = se.values
```
